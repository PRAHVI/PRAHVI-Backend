```python
paper_sources = ['http://thehill.com/', 
                 'http://www.boston.com/', 
                 'http://edition.cnn.com/', 
                 'https://www.bostonglobe.com/',
                 'http://www.theblaze.com/',
                 'http://www.msnbc.com/',
                 'http://www.huffingtonpost.com/',
                 'http://www.theatlantic.com/world/',
                 'http://www.reuters.com/',
                 'http://www.foxnews.com/',
                 'https://mic.com/',
                 'http://www.chicagotribune.com/',
                 'http://www.theverge.com/',
                 'http://www.bbc.com/news',
                 'http://www.cnbc.com/',
                 'http://www.ap.org/',
                 'http://www.salon.com/',
                 'http://nypost.com/',
                 'http://www.vice.com/en_us',
                 'http://www.usatoday.com/',
                 'http://www.examiner.com/',
                 'http://www.economist.com/',
                 'http://www.newsweek.com/',
                 'http://www.bloomberg.com/',
                 'http://www.businessinsider.com/',
                 'http://www.cbsnews.com/',
                 'http://www.breakingnews.com/',
                 'http://theweek.com/',
                 'http://www.breitbart.com/',
                 'http://www.thedailybeast.com/',
                 'http://www.usnews.com',
                 'http://www.slate.com/',
                 'http://www.npr.org/',
                 'http://www.nytimes.com/',
                 'http://www.nydailynews.com/news/money',
                 'https://www.washingtonpost.com/',
                 'http://www.nbcnews.com/',
                 'http://www.ft.com/',
                 'https://www.upworthy.com/',
                 'http://www.vox.com/',
                 'http://www.wsj.com/',
                 'http://www.forbes.com/',
                 'https://www.buzzfeed.com/',
                 'http://bleacherreport.com/',
                 'http://www.aol.com/',
                 'http://time.com/',
                 'http://gawker.com/',
                 'http://www.latimes.com/',
                 'http://fortune.com/',
                 'http://abcnews.go.com/']

hashes = set()
idf = {}
num_documents = 0

for source in paper_sources:
    paper = newspaper.build(source, language='en')
    
    print 'Processing articles for source: {}'.format(source)
    for article in paper.articles:
        try:
            article.download()
            article.parse()

            if not hasattr(article, 'link_hash') or article.link_hash not in hashes:
                uniq_words = {uniq_word for uniq_word in article.text.split() if uniq_word != ''}
                num_documents += 1

                for word in uniq_words:
                    if idf.get(word):
                        idf[word] += 1
                    else:
                        idf[word] = 1
        except:
            continue
    
```

```python
import numpy as np
import os
import cv2
import json


idf_data = json.load(open('data.json', 'rb'))
num_doc = idf_data['num_documents']
idfs = idf_data['idf']



class TFIDF:
    """Class to compute the TFIDF of a document"""
    def __init__(self, text):
        text = text.split(' ')
        
        result = {}


        tf = {}
        for word in text:
            if word == '':
                continue
            if tf.get(word):
                tf[word] += 1
            else:
                tf[word] = 1

        max_tf = max(tf.values())

        for word in tf.keys():
            tf[word] = 0.5 + 0.5 * (tf[word] / max_tf)
            idf_val = np.log(float(num_doc) / idfs[word])
            result[word] = tf[word] * idf_val

        self.result = result
```

```python
sample = 'How to use the Study Groups effectively'
t = TFIDF(sample)
print t.result
```
